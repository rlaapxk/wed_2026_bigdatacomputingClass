# 1. 넘파일 다차원 배열(ndarray) 객체 생성
- import numpy as np
- 리스트을 이용하여 다차원 배열 생성
- np.array(리스트)
  - 리스트 → 배열로 바꾸고 싶을 때
  - 단순 저장 → 수학 연산이 필요할 때
  - 데이터 분석/머신러닝 모델에 데이터를 입력할 때
- 속성 :
  - ndim : 배열 축 혹은 차원의 개수
  - shape : 배열의 형상을 (n, m) 형식의 튜플형으로 표시
  - size :
    - 배열 원소의 개수
    - (m,n) 형상 배열의 size = m*n
  - dtype : 배열 원소의 자료형을 기술
  - astype : 배열 원소 형 변환

```python
# 리스트를 이용한 1차원 배열 생성
import numpy as np
data  = [10, 20, 30]
arr = np.array(data) # 1차원 배열 반환

print("data:",data, type(data))
print("arr:",arr, type(arr))
print("="* 50)
print("arr.ndim:", arr.ndim)
print("arr.shape:", arr.shape)
print("arr.size:", arr.size)
print("arr.dtype:", arr.dtype)
```

```python
# 리스트를 이용한 2차원 배열 생성
# 머신러닝에서 행은 샘플의 개수, 열은 각 샘플이 지니고 있는 특징의 개수를 표현

data = [[90, 80, 90], [70, 80, 90]]
data_2d = np.array(data)
print("data_2d:", data_2d)
print("="* 50)
print("data_2d.ndim:", data_2d.ndim)
print("data_2d.shape:", data_2d.shape)
print("data_2d.size:", data_2d.size)
print("data_2d.dtype:", data_2d.dtype)
```

```python
# 3차원 넘파이 배열을 이용해서 이미지 데이터 기록
# 3차원 배열은 (axis0, axis1, axis2) = (depth, height, width)
# 디지털 이미지 데이터 (가로, 세로, RGB 채널)

# 2x2 크기의 컬러 이미지 하나
tensor_list = [
    [[255, 0, 0], [0, 255, 0]],
    [[0, 0, 255], [255, 255, 0]]
    ]

tensor = np.array(tensor_list)
print(tensor)
print("="* 50)
print("tensor.ndim:",tensor.ndim)
print("tensor.shape:",tensor.shape)
print("tensor.size:",tensor.size)
print("tensor.dtype:",tensor.dtype)
```

# 2. 넘파이 다차원 배열 간 사칙 연산
  - 피연산자 다차원 배열들의 형상이 같다는 가정
  - +, -, *, / 연산 수행
  - 배열 프로그래밍
  - 선형대수의 벡터화 연산
  - 벡터을 병렬 연산 수행
    - 데이터 병렬 처리

```python
# 1차원 배열 사칙연산 -> 1차원 배열 결과로 반환

# 1차원 넘파이 배열 생성
store_a = [1000, 2000, 1500]
store_b = [2000, 1000, 1000]

arr_a = np.array(store_a)
arr_b = np.array(store_b)

# 형태가 같은 1차원 배열 간 연산
print(arr_a + arr_b)
print(arr_a - arr_b)
print(arr_a * arr_b)
print(arr_a / arr_b)
```

```python
# 2차원 배열 사칙연산: 동일한 형태를 가지는 배열(2x3)
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr2_2d = np.array([[10, 20, 30], [40, 50, 60]])

# 형태가 같은 2차원 배열 간 연산
print(arr_2d + arr2_2d,"\n")
print(arr_2d - arr2_2d,"\n")
print(arr_2d * arr2_2d,"\n")
print(arr_2d / arr2_2d,"\n")
```

# 6. 다차원 배열의 지정된 축 방향으로 원소들을 뒤집기
  - np.flip(배열객체, axis=)

```python
# (4, 5) 구조의 이미지 데이터를 생성 -> 상하, 좌우 반전시켜 새로운 이미지 만들기

import numpy as np
# 데이터 생성
img = np.arange(20).reshape((4, -1))
print(img)
print("\n")
# (1) 좌우 반전 (axis = 1)
h_flip = np.flip(img, axis=1)
print(h_flip)
print("\n")
# (2) 상하 반전 (axis = 0)
v_flip = np.flip(img, axis=0)
print(v_flip)
print("\n")
# (3) 상하좌우 전체반전
all_flip = np.flip(img)
print(all_flip)
print("\n")
```

# 7. 넘파이 다차원 배열의 인덱싱
- 다차원 배열의 특정 원소 추출 방식
- 인덱스 정수을 이용
- 인덱싱은 두 축을 이용해서 함
  - axis 0 ; 행 참조
  - axis 1 ; 열 참조
- arr[][]
- arr[,]
- arr[]


```python
# 3차원 배열 슬라이싱
import numpy as np

#데이터 생성: 60개의 원소로 구성된 (3, 4, 5) 형태의 3차원 배열 생성 (3층, 4행, 5열)
tensor = np.arange(60). reshape((3, 4, 5))
print(tensor)
print("=" * 65)
print("=" * 65)

# (1) 특정 층 전체 가져오기 (첫번째 층)
print(tensor[0, :, :])
print("=" * 65)

# (2) 모든 층에서 특정 행만 가져오기 (모든 층의 첫번째 행)
print(tensor[:, 0, :])
print("=" * 65)

# 모든 층에서 왼쪽 벽면 (첫번째 열)만 수직으로 잘라내기
print(tensor[:, :, 0])
print("=" * 65)

# (4) 정가운데 부분만 큐브 형태로 도려내기
print(tensor[1:3, 1:3, 1:4]) # 1~2층의 1~2행에서 1~3열
print("=" * 65)
```

# 9.  넘파이 다차원 배열의 슬라이싱을 이용한 논리인덱싱
- 특정한 조건을 만족하는 배열의 원소값만을 추출하여 1차원 배열 객체로 반환
- 브로드캐스팅하여 병렬 연산 처리


1. 주어진 배열을 이용하여 조건식을 만족하는 동일한 형상을 가지는 불리언 배열을 생성

2.   주어진 배열에 불리언 배열을 인덱스로 지정

- cf. np.where(조건식, x, y)




```python
# 2차원 배열 데이터 생성
arr = np.arange(100).reshape(5, -1) # (5, 20)
print(arr, "\n")

# ex1) 주어전 배열에서 짝수들만 추출하여 1차원 배열로 생성
# (1) 불리언 배열
arr_bool = arr % 2 == 0
print(arr_bool, "\n")

# (2) 불리언 인덱싱
arr_even = arr[arr_bool]
print(arr_even)

# ex2) 주어진 배열에서 20보다 큰 요소들만 추출하여 1차원 배열로 생성
arr_greater = arr[arr > 20]
print(arr_greater)

# ex3) 주어진 배열애서 20보다 큰 요소들만 추출
arr_filter = np.where(arr > 20, arr, 0)
print(arr_filter)
```

```python
# 추가 수익률 계산 프로그램(평균 0.01, 표준편차 0.02인 난수 1000개)
# 의미: 이런 문제의 불확실성을 난수로 생성하여 향후 1000일 간의 주가를 예측해보자

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import warnings
import os

# 1. 환경 설정 (경고 차단 및 한글 폰트)
warnings.filterwarnings('ignore')
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
if not os.path.exists(font_path):
    !sudo apt-get install -y fonts-nanum > /dev/null
font_prop = fm.FontProperties(fname=font_path)


# 2. 데이터 준비
np.random.seed(42) # 재현성 보장
daily_values = np.random.normal(0.001, 0.02, 1000)
price = np.cumprod(1 + daily_values) # price.max(),price.min()
# print(price)

# simulation : line graph
plt.figure(figsize=(14,5))
plt.plot(np.arange(1000), price)
plt.title("주가 수익률 시뮬레이션", fontproperties = font_prop, fontsize = 15)
plt.xlabel("일수(days)",fontproperties = font_prop, fontsize = 13)
plt.ylabel("주가(price)", fontproperties = font_prop, fontsize = 13)
plt.xticks(np.arange(0,1001,100))
plt.legend()
plt.show()
```

# 14. 다차원 배열 프로그래밍
- 벡터화 연산 (데이터병렬 처리)
  - 배열 간 연산
  - 리덕션
    - np.sum(arr, axis=)
    - np.mean(arr, axis=)
    - arr.sum(axis=0)
    - arr.sum(axis=1)
    - arr.sum(axis=(0,1))
  - 조건 리덕션 원리 (Boolean to Integer)
      - NumPy에서 리덕션이 강력한 이유는 **논리값(True/False)**을 연산하는 순간 **숫자(1/0)**로 자동 변환.

      - np.sum(condition) : 조건이 참(True)인 개수를 구함

      - np.mean(condition) : 조건이 참(True)인 비율을 구함 (예: 0.7이면 70%가 참)
      - 이 리덕션 기법들을 활용해서 실제 데이터셋에서 "이상치(Outlier)"를 찾기

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import warnings
import os

warnings.filterwarnings('ignore')
font_path = '/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf'
if not os.path.exists(font_path):
    !sudo apt-get install -y fonts-nanum > /dev/null
font_prop = fm.FontProperties(fname=font_path)


np.random.seed(42)
scores = np.random.randint(60, 100, 12).reshape(3, -1)
supject = ["국어", "영어", "수학", "통계"]
student = ["영희", "철수", "신한"]

subject_avg = np.mean(scores, axis=0).round(2)
student_avg = np.mean(scores, axis=1).round(2)

plt.figure(figsize=(20,6))
plt.suptitle("원본 데이터 리덕션 데이터 분석", fontproperties = font_prop, fontsize = 20)

plt.subplot(1, 3, 1)
im = plt.imshow(scores, cmap = "YlGnBu", interpolation = "nearest")
plt.title("원본 성적 분포", fontproperties = font_prop, fontsize = 15)
plt.xlabel("과목명", fontproperties = font_prop)
plt.ylabel("학생명", fontproperties = font_prop)
plt.xticks(np.arange(len(supject)), supject, fontproperties = font_prop)
plt.yticks(np.arange(len(student)), student, fontproperties = font_prop)
cbar = plt.colorbar(im)
cbar.set_label("점수", fontproperties = font_prop)
for i in range(3):
  for j in range(4):
    plt.text(j, i, scores[i, j], ha = "center", va = "center", color = "black")


plt.subplot(1, 3, 2)
plt.bar(supject, subject_avg, color = ['red', 'green', 'blue', 'yellow'], edgecolor = 'black')
plt.title("과목별 평균", fontproperties = font_prop, fontsize = 15)
plt.xlabel("과목명", fontproperties = font_prop)
plt.ylabel("평균 점수", fontproperties = font_prop)
plt.xticks(np.arange(len(supject)), supject, fontproperties = font_prop)
for i , score_val in enumerate(subject_avg):
  plt.text(i, score_val + 0.5, str(score_val),  ha = "center", va = "bottom", color = "black", fontsize = 12)


plt.subplot(1, 3, 3)
plt.barh(student, student_avg, color = "lightgreen", edgecolor = 'black')
plt.title("학생별 평균 분포", fontproperties = font_prop, fontsize = 15)
plt.xlabel("평균 점수", fontproperties = font_prop)
plt.ylabel("학생명", fontproperties = font_prop)
plt.yticks(np.arange(len(student)), student, fontproperties = font_prop)
for i , score_val in enumerate(student_avg):
  plt.text(score_val + 1, i, str(score_val), va = "center", color = "black", fontsize = 12, fontweight = 'bold')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

# 15. 넘파일 배열의 결합
- np.concatenate(튜플)
- np.vstack(튜플 또는 리스트)
- np.hstack(튜플 또는 리스트)
- np.r_[ 콤마 사용하여 나열 ]
- np.c_[콤마 사용하여 나열]
- np.column_stack(리스트)


```python
# ex1 1차원 배열간 결합

import numpy as np
# 데이터 생성
a = np.array([1, 2, 3]) # (3, )
b = np.array([4, 5, 6]) # (3, )

# 세로 쌓기 (수직 방향) -> 행 추가해서 2차원 배열
res_stack = np.vstack((a, b))
print(res_stack) # (2, 3)

# 가로로 붙이기 (수평방향) -> 열 추가해서 1차원 배열
res_hstack = np.hstack((a, b))
print(res_hstack)
res_concat = np.concatenate((a, b), axis = 0)
print(res_concat)

# 주의: column_stack() -> 2차원 배열
res_col = np.column_stack((a, b))
print(res_col)

# EX2 형태가 다른 배열간 결합

arr_1 = np.arange(1, 4)
arr_2 = np.arange(4, 10).reshape(2, 3)

# 방법1
# (1) arr_1 -> (1,3) - 열의 개수를 일치
arr_1_2d = arr_1.reshape(-1, 3)
print(arr_1_2d)
combined_2d = np.concatenate((arr_1_2d, arr_2), axis = 0)
print(combined_2d)

# 방법 2
v_combined = np.vstack((arr_1, arr_2))
print(v_combined)
```
